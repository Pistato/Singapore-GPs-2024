```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, WebDriverException
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.singhealth.com.sg/rhs/GP-Result?ClinicName=&ProgrammesAndInitiatives=CHAS&OpenHours=allGPs&Location="

begin = time.time()

options = Options()
options.headless = True
options.add_argument('--log-level=3')
driver = webdriver.Chrome(options=options)
```

```python
# Create an empty DataFrame
clinic_df = pd.DataFrame(columns=['Clinic Name', 'Distance and Status', 'Address', 'Location', 'Contact Info'])

next_page_no = 1
driver.get(url)
# html_text = driver.page_source
# soup = BeautifulSoup(html_text)
while(next_page_no <= 214):
    # if next_page_no <= 211:
    #     print(str(next_page_no))
    #     if (next_page_no % 10 == 1):
    #         print("Next 10 pages")
    #         xpath_expression = f"//a[contains(@href, 'Page${next_page_no}')]"
    #     else:
    #         xpath_expression = "//a[text()='" + str(next_page_no) + "']"
    #     element = driver.find_element("xpath", xpath_expression)

    #     element.click()
    #     WebDriverWait(driver, 6)
    #     next_page_no = next_page_no + 10
    #     continue
    if (next_page_no % 20 == 1):
        # File path to save the CSV file
        csv_file_path = "GP" + str(next_page_no // 20) +".csv"

        clinic_df.to_csv(csv_file_path, index=True)
        # Clear DataFrame
        clinic_df = pd.DataFrame(columns=['Clinic Name', 'Distance and Status', 'Address', 'Location', 'Contact Info'])

    next_page_no = next_page_no + 1
    html_text = driver.page_source
    soup = BeautifulSoup(html_text) 
    
    # Find all div elements with class 'sh-detail-contact-item'
    clinic_divs = soup.find_all("div", class_="sh-detail-contact-item col-md-6 col-xs-12 section_main-content")

    # Extract information from each div
    for clinic_div in clinic_divs:
        clinic_name = clinic_div.find("label", class_="sh-detail-contact-item-title").text.strip()
        distance_and_status = clinic_div.find("label", class_="sh-detail-contact-item-distance").text.strip()
        address = clinic_div.find_all("a")[0].text.strip()
        address = address.replace('\n', '')
        location = clinic_div.find_all("a")[1].text.strip()
        contact_info = clinic_div.find("div", class_="sh-detail-contact-item-phone").text.strip()
        contact_info = contact_info.replace('\n', ' ')
        
        print("Clinic Name:", clinic_name)
        print("Distance and Status:", distance_and_status)
        print("Address:", address)
        print("Location:", location)
        print("Contact Info:", contact_info)
        print()
        data_row = [clinic_name, distance_and_status, address, location, contact_info] 
        clinic_df.loc[len(clinic_df)] = data_row

    print(str(next_page_no))
    if (next_page_no % 10 == 1):
        print("Next 10 pages")
        xpath_expression = f"//a[contains(@href, 'Page${next_page_no}')]"
    else:
        xpath_expression = "//a[text()='" + str(next_page_no) + "']"
    element = driver.find_element("xpath", xpath_expression)

    element.click()
    WebDriverWait(driver, 6)
```

```python
# File path to save the CSV file
csv_file_path = "GP11.csv"

clinic_df.to_csv(csv_file_path, index=True)

print(f"Data has been written to '{csv_file_path}' successfully.")
```

```python
import pandas as pd

# List of file names or paths
file_names = ['GP1.csv', 'GP2.csv', 'GP3.csv', 'GP4.csv', 'GP5.csv', 'GP6.csv', 'GP7.csv', 'GP8.csv', 'GP9.csv', 'GP10.csv', 'GP11.csv']

# Create an empty list to store DataFrames
dfs = []

# Iterate through each file and read it into a DataFrame, then append to the list
for filename in file_names:
    fp = "./save/" + filename
    dfs.append(pd.read_csv(fp))

# Concatenate all DataFrames in the list along the rows
combined_df = pd.concat(dfs, ignore_index=True)

# Optionally, you can reset the index if you want consecutive row indices
# combined_df.reset_index(drop=True, inplace=True)
combined_df.drop(columns=['Unnamed: 0'], inplace=True)
# Now you have a single DataFrame containing data from all CSV files
combined_df.to_csv("All GPs.csv", index=True)
```
