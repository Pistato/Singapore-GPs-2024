```python
import pandas as pd
```

```python
all_gp = pd.read_csv("..\\2. CoordinatesFromPostalCode\\Output Institution\\All GPs.csv", index_col=0)
all_hospital = pd.read_csv("..\\2. CoordinatesFromPostalCode\\Output Institution\\All Hospitals.csv", index_col=0)
all_poly = pd.read_csv("..\\2. CoordinatesFromPostalCode\\Output Institution\\All Polyclinics.csv", index_col=0)
locations = pd.read_csv("..\\2. CoordinatesFromPostalCode\\Output Institution\\combined_postal_codes.csv")
```

```python
from geopy.distance import geodesic
```

```python
# remove all rows where Lat is null or blank
all_gp = all_gp[all_gp["Lat"].notnull()]
```

```python
# Function to calculate distance between two points using geodesic distance
def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).kilometers

# Initialize empty dataframes to store results
closest_gp = pd.DataFrame(columns=['PostalCode', 'ClosestGP', 'GPDistance'])
closest_hospital = pd.DataFrame(columns=['PostalCode', 'ClosestHospital', 'HospitalDistance'])
closest_polyclinic = pd.DataFrame(columns=['PostalCode', 'ClosestPolyclinic', 'PolyclinicDistance'])

# Iterate over each row in the locations dataframe
for index, location in locations.iterrows():
    print("Processing index: " + str(index))
    pc = location['Postal Code']
    lat1, lon1 = location['Latitude'], location['Longitude']
    coord1 = (lat1, lon1)
    
    # Calculate distances to all general practitioners
    gp_distances = [(calculate_distance(coord1, (gp['Lat'], gp['Long'])), gp['Clinic Name']) for _, gp in all_gp.iterrows()]
    closest_gp_distance, closest_gp_name = min(gp_distances)
    closest_gp = pd.concat([closest_gp, pd.DataFrame({'PostalCode': [pc], 'ClosestGP': [closest_gp_name], 'GPDistance': [closest_gp_distance]})], ignore_index=True)
    
    # Calculate distances to all hospitals
    hospital_distances = [(calculate_distance(coord1, (hospital['Lat'], hospital['Long'])), hospital['Clinic Name']) for _, hospital in all_hospital.iterrows()]
    closest_hospital_distance, closest_hospital_name = min(hospital_distances)
    closest_hospital = pd.concat([closest_hospital, pd.DataFrame({'PostalCode': [pc], 'ClosestHospital': [closest_hospital_name], 'HospitalDistance': [closest_hospital_distance]})], ignore_index=True)
    
    # Calculate distances to all polyclinics
    polyclinic_distances = [(calculate_distance(coord1, (polyclinic['Lat'], polyclinic['Long'])), polyclinic['Clinic Name']) for _, polyclinic in all_poly.iterrows()]
    closest_polyclinic_distance, closest_polyclinic_name = min(polyclinic_distances)
    closest_polyclinic = pd.concat([closest_polyclinic, pd.DataFrame({'PostalCode': [pc], 'ClosestPolyclinic': [closest_polyclinic_name], 'PolyclinicDistance': [closest_polyclinic_distance]})], ignore_index=True)


# Display the results
print("Closest General Practitioners:")
print(closest_gp)
print("\nClosest Hospitals:")
print(closest_hospital)
print("\nClosest Polyclinics:")
print(closest_polyclinic)
```

```python
closest_gp.to_csv("Closest GP.csv")
closest_hospital.to_csv("Closest Hospital.csv")
closest_polyclinic.to_csv("Closest Polyclinic.csv")
```
